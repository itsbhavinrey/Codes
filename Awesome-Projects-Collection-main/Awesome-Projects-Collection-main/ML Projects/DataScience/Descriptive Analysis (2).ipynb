```python
import matplotlib.pyplot as plt
from prettytable import PrettyTable
from tabulate import tabulate
import numpy as np
df = pd.DataFrame(file)
outlier = []
count = 0
t = PrettyTable(['Column', 'Outlier'])
for i in df.columns:
    count = 0
    lq = df[i].quantile(.25)
    hq = df[i].quantile(.75)
    IQR = hq-lq
    md = df[i].mean()
    if(i != "Class"):
        for j in range(len(df[i])):
            if(df[i][j] < lq - 1.5*IQR or df[i][j] > hq + 1.5*IQR):
                df[i][j] = md
                count += 1
    t.add_row([i,count])
    outlier.append(count)
ii = 0
print(t)
for i in df.columns:
    plt.boxplot(x=np.array(df[i]))
    plt.show()
t1 = PrettyTable(['Column', 'Outlier'])   
outlier1 = []
for i in df.columns:
    count = 0
    lq = df[i].quantile(.25)
    hq = df[i].quantile(.75)
    IQR = hq-lq
    md = df[i].median()
    if(i != "Class"):
        for j in range(len(df[i])):
            if(df[i][j] < lq - 1.5*IQR or df[i][j] > hq + 1.5*IQR):
                count += 1
    t1.add_row([i,count])
    outlier1.append(count)
print(t1)
```
